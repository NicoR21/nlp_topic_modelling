# Results comparison

### Merge all needed variables generated from previous notebooks

```python
def merge_column_results(df_LDA: pd.DataFrame, df_OT: pd.DataFrame, df_SA: pd.DataFrame) -> pd.DataFrame:
    final_df = pd.merge(df_LDA, df_OT, left_index=True, right_index=True)
    final_df = pd.merge(final_df, df_SA, left_index=True, right_index=True)
    final_df = final_df[["Review_x", "Liked_x", POLARITY, SENTIMENT_BINARY, SENTIMENT_TYPE, 
                         CLUSTER_OT, DOMINANT_TOPIC_LDA, TOPIC_PERCENTAGE_CONTRIBUTION, KEYWORDS]]
    final_df.columns = [REVIEW, LIKED, POLARITY, SENTIMENT_BINARY, SENTIMENT_TYPE, 
                        CLUSTER_OT, DOMINANT_TOPIC_LDA, TOPIC_PERCENTAGE_CONTRIBUTION, KEYWORDS]
    return final_df
```

```python
df_comparison = merge_column_results(df_cluster_LDA, df_cluster_OT, df_sentiment_analysis)
df_comparison.head()
```

### Compare sentiment_type between optimal transport and LDA methods

```python
def display_clutsers_sentiment(df: pd.DataFrame, cluster_var: str) -> dict:
    df_count = pd.crosstab(df[cluster_var], df[SENTIMENT_TYPE], margins=True, margins_name="Total")
    df_percent = pd.crosstab(df[cluster_var], df[SENTIMENT_TYPE], normalize="index")
    return {"DF_COUNT": df_count,"DF_PERCENT": df_percent}
```

```python
display_clutsers_sentiment(df_comparison, CLUSTER_OT)["DF_PERCENT"]
```

```python
display_clutsers_sentiment(df_comparison, DOMINANT_TOPIC_LDA)["DF_PERCENT"]
```
